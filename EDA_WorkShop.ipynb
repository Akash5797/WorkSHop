#### EDA(Explolatery Data Analysis) is the process of analyzing and visualizing a dataset to understand its characteristics before applying any machine learning models or statistical techniques   

* It helps identify patterns, relationship, missing values, and anomalies in the data.
* `Python Libraries`: Pandas Matplotlib, Seaborn, Plotty

#### Key Steps in EDA -
* `Understanding the Data` - Load the dataset and check its structure using `.head(), .info() & .describe()`.
* `Handling Missing values` - Indetify and treat missing values using `isnull(), sum()`.
* `Checking Data Types` - Ensure Data Types are appropriate for analysis.
* `Indetifying Outliers` - Use Box plots to detect anmalies.
* `Visulaizing Data` - Use `Histogram, Scatter Plots, Heatmaps, & Pair Plots` to explore relationships.
* `Feature Correlation` - Analyze how features are related using correlation matrices.
* `Feature Engineering` - Create new meaningful features from exiting ones.

```python
import ollama  # Importing the ollama library to use local AI models for generating insights

# Define a function to generate insights from a dataset summary
def generate_insights(DS_summary):
    # Create a prompt by combining a message with the dataset summary
    prompt = f"Analyze the Dataset Summary and provide key insights:\n\n{DS_summary}"
    
    # Call the ollama.chat function with the chosen model ('mistral') and the prompt
    response = ollama.chat(
        model='mistral',  # Specify which model to use (here it's 'mistral')
        messages=[{"role": "user", "content": prompt}]  # Pass the prompt as a user message
    )
    
    # Extract and return only the content part of the AI's response
    return response['message']['content']

# Generate a descriptive statistical summary of the dataset (DS)
summary = DS.describe().to_string()  # Converts the summary into a string format for the AI to read

# Call the function with the summary and store the AI-generated insights
insights = generate_insights(summary)

# Print a title for clarity in the output
print("\nAI-GENERATED INSIGHTS:\n")

# Print the AI-generated insights to the console
print(insights)
```

```python
import gradio as gr
```

```python
# Function for EDA analysis using uploaded file
def eda_analysis(file):
    DS = pd.read_csv(file)                      # Read CSV file into a DataFrame
    summary = DS.describe().to_string()         # Get descriptive statistics summary of dataset
    insights = generate_insights(summary)       # Generate AI-based insights from the summary
    return insights

# Create a Gradio interface for the EDA analysis
EDA = gr.Interface(
    fn=eda_analysis,          # Function to call when user uploads a file
    inputs='file',            # Accept file input from user
    outputs='text'            # Display insights as plain text
    title="LLM-Powered EDA Data Analyzer Created By Akash "
)

# Launch the Gradio interface and allow sharing via public link
EDA.launch(share=True)
```
